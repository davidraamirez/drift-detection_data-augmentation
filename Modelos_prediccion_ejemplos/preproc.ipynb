```python
import pandas as pd
```

```python
df = pd.read_csv("updated_pollution_dataset.csv")
```

```python
df1=df.drop('Air Quality', axis=1)
```

```python
df1
```

```python
def series_periodos(inicio, periodos, freq): 
    serie = pd.date_range(start=inicio, periods=periodos, freq=freq)
    return serie
```

```python
indice=series_periodos('1/1/2010',df1.shape[0],'D')
```

```python
df1.index=indice
```

```python
df1.to_csv("Preprocesado.csv",index_label="indice")
```
