RELACIÓN LINEAL:

```python
# Crea una columna Target: y = a + b * x
def objetivo_lineal(df_caract,a,b):
    df = df_caract.copy()
    df['Target'] = a + b * df_caract[df_caract.columns[0]]
    return df
```

```python
objetivo_lineal(df,2,3)
```

RELACIÓN NO LINEAL:

```python
# Crea una columna Target: y = sumatorio a[i] * x ^ i

def objetivo_polinómico(df_caract,a):
    df = df_caract.copy()
    df['Target'] = np.zeros(df.shape[0])
    for i in range(0,len(a)):
        df['Target'] = df['Target'] + a[i]*df_caract[df_caract.columns[0]]**i
    return df
```

```python
objetivo_polinómico(df,[1,2,3])
```

```python
# Crea una columna Target: y = a * e ^ (b * x)

def objetivo_exp(df_caract,a,b):
    df = df_caract.copy()
    df['Target'] = a * np.exp(b* df_caract[df_caract.columns[0]])
    return df
```

```python
objetivo_exp(df,4,0.1)
```

```python
# Crea una columna Target: y = a + b * log (x)

def objetivo_log(df_caract,a,b):
    df = df_caract.copy()
    df['Target'] = a + b * np.log(df_caract[df_caract.columns[0]])
    return df
```

```python
objetivo_log(df,10,5)
```

MULTIVARIANTE

```python
# Crea una columna Target: y = a + sumatorio b[i] * x [i]
def multivariante(df_caract,a,b):
    df = df_caract.copy()
    df['Target'] = a
    for k in range(0,len(b)):
        df['Target'] = df['Target'] + b[k] * df_caract[df.columns[k]]
    return df
```

```python
multivariante(df2,70,[0.5,0.7])
```

INTERACCIÓN:

```python
# Crea una columna Target: y = a + sumatorio b[i][i] * x [i] + sumatorio b[j][k]x[j]x[k]

def interaccion(df_caract,a,b):
    df = df_caract.copy()
    df['Target'] = a
    for k in range(0,b.shape[0]):
        df['Target'] = df['Target'] + b[k][k] * df_caract[df.columns[k]]
        for i in range(k+1,b.shape[1]):
            df['Target'] = df['Target'] + b[k][i] * df_caract[df.columns[k]] * df_caract[df.columns[i]]
    return df
```

```python
interaccion(df1,-25,np.array(([0.5,0.25,0.65],[0.25,0.3,0.6],[0.65,0.6,0.1])))
```

PROPORCIONALIDAD INVERSA:

```python
# Crea una columna Target: y = a / x ^ n

def objetivo_prop_inversa(df_caract,a,n):
    df = df_caract.copy()
    df['Target'] = a / (df_caract[df_caract.columns[0]] ** n)
    return df
```

ESCALONADA:

```python
# Crea una columna Target: Si x < umbral, y = f(x). En otro caso, y = g(x)

def objetivo_escalonada(df_caract,f,g,umbral):
    df = df_caract.copy()
    df['Target'] = np.zeros(df_caract.shape[0])
    for k in df.index:
        x = df_caract.loc[k,df.columns[0]]
        if x < umbral:
            df.loc[k,'Target'] = f(x)
        else :
            df.loc[k,'Target'] = g(x)     
    return df
```

```python
def f(x):
    return x**3
```

```python
def g(x):
    return x-7
```

```python
objetivo_escalonada(df,f,g,52)
```

```python
# Crea una columna Target: Si x cumple condM --> y = fM(x1,...,xN)

def objetivo_condicional(df_caract,cond,func):
    df = df_caract.copy()
    df['Target'] = np.zeros(df_caract.shape[0])
    for k in df_caract.index:
        a = list()
        for j in df_caract.columns:
            a.append(df_caract.loc[k,j])
        
        for i in range(0,len(cond)):
            if cond[i](a):
               df.loc[k,'Target'] = func[i](a)
    return df
```

```python
def cond1(x):
    return x[0] > x[1]

def cond2(x):
    return x[0]== x[1]

def cond3(x):
    return x[0] < x[1]
```

```python
def f1(x):
    return x[0] - x[1]

def f2(x):
    return x[0]/2

def f3(x):
    return x[1] - x[0]
```

```python
objetivo_condicional(df2,[cond1,cond2,cond3],[f1,f2,f3])
```

```python
# Crea una columna Target: y = f(x1,...,xN)

def objetivo_funcional(df_caract,f):
    df = df_caract.copy()       
    df['Target'] = np.zeros(df_caract.shape[0])
    for k in df_caract.index:
        a = list()
        for j in df_caract.columns:
            a.append(df_caract.loc[k,j])
        df.loc[k,'Target']=f(a)
    return df
```
